Additional Insight: 

To better understand the consistency of engagement, both variance and interquartile range (IQR) were considered. Variance captures the overall spread of the data, including extreme values, while IQR focuses on the middle 50% of observations and is less influenced by outliers.In this analysis, the control group showed greater variability, indicating less consistent engagement. The treatment group, however, demonstrated more concentrated viewing times, suggesting a more stable user experience.
Given the presence of outliers, IQR provided a more reliable view of typical engagement behavior. This reinforces the conclusion that the redesigned slides not only increased average viewing time but also improved consistency across users.
